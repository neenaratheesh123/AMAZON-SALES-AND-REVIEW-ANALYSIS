## Final statistics for insight report

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('amazon_cleaned.csv')
```

```python
summary ={
    "Total Products"          : len(df),
    "Avg Rating"              : round(df['rating'].mean(), 2),
    "Avg Discount %"          : round(df['discount_percentage'].mean(), 1),
     "Avg Price"              : round(df['discounted_price'].mean(), 0),
    "Avg Savings"             : round(df['savings_amount'].mean(), 0),
     "% Rated >= 4.0"         : round((df['rating']>= 4.0).mean()*100, 1),
    "Corr : Price vs ActualpPrice": round(df['discounted_price'].corr(df['actual_price']), 3),
    "Corr :Price vs Rating"   : round(df['discounted_price'].corr(df['rating']), 3)
}
print("="*50)
print("   FINAL SUMMARY STATISTICS")
print("="*50)
for k, v in summary.items():
    print(f"  {k:<28}: {v}")
    print("="*50)
   
```

  * ## Significant Patterns, Correlations

#### Anomaly - Extreme discounts signal low quality**
poor-rated Budget products carry the highest average discount (-64.6%)
compared to -47.8% for Excellent- rated Budget products. This suggests
sellers use deep discounting to clear low-quality stock.
#### Pattern- Bimodal discount distribution**
 Discounts cluster around 40% and 60% rather than spreading evenly -
a sign of psychological pricing anchors used by sellers.
 #### Correlation - Price is structurally honest**
 price and actual price correlate at r=0.96,meaning
 MRP is notartificially inflated to fake bigger discounts.
#### Pattern - Category specialization**
 -- Electronics- dominates Luxury tier, highest absolute savings 
 -- Computers & Accessories-best value(high discount + high rating)
 -- Office products-highest rating,lowest discount(quality niche)
                                     
 

* ## 5 Key Insights

```python
insights =[
    ("Insight 1", "Amazon India is a Quality-Positive Marketplace",
     f"{summary['% Rated >= 4.0']}% of products rated >= 4.0 stars."
     "Below-average products are a clear minority."),
    
    ("Insight 2", "Extreme Discounts Signal Lower Quality",
    "Products with >=70% discount average a lower rating than those"
    "with <40% discount.Treat huge discounts as a red flag." ),

    ("Insight 3", "Computer & Accessories =Best value Category",
     "Highest avg discount(-54% AND a strong  avg rating(-4.16)-"
    "the best balance of price-cutting and quality."),

    ("Insight 4", "Price Does Not Predict Rating",
     f"Correlation(price, rating) = {summary["Corr :Price vs Rating" ]} "
     "- expensive items are not meaningfully better rated."),

     ("Insight 5", "Review Volume Reflects  Purchase Frequency, Not Quality",
     "Top reviewed products are  cheap , frequently-bought consumables"
      "(cables, adapters) - not premium items."),
]
print("="*60)
print(" KEY INSIGHTS")
print("="*60)
for num,title, desc in insights:
    print(f" {num}: {title}")
    print(f"  {desc}")
 
     
```

 * ### Conclusion  and  Recommendations

#Overall Findings*
    
  The Amazon India dataset reflects a quality- positive marketplace where customer ratings are reliable ,pricing is strucyurally honest, and discount
depth does not guarantee value.
    
 #Recommendations*

For Consumers:
    
- Prioritize ratings/reviews over discount percentage 
- Treat discounts >= 70% with caution
- Computers & Accessories offers the most reliable value across price tiers
    
For Sellers:
    
- Quality first positioning(low discount ,high rating) is viable - see Office Products
- Avoid Extreme discounting: it correlates with lower perceived quality
- Mid-range pricing (Rs 300-999) has the highest competition and volume
